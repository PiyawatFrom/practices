```python
import pandas as pd
df_mk_camp = pd.read_csv("gdrive/MyDrive/Data Master Class/Week 2 Tools for Data Sci/Week 2 material and code/week02/marketing_campaign.csv" ,sep='\t')
df_mk_camp
```

```python
df_mk_camp.info()
```

พบว่าใน Income มี null อยู่ แต่เพราะไม่ได้จะพิจารณาช่องนี้ จึงปล่อยไว้

```python
df_CusInfo = df_mk_camp[ [ "ID" , "Income" , "MntMeatProducts","MntFishProducts" , "MntFruits" ,"MntSweetProducts","MntWines" , "Marital_Status" ,"Kidhome" ,"Teenhome"] ]
#ดึงมาเฉพาะข้อมูลที่น่าจะเกี่ยวกับสุขภาพและครอบครัว
```

```python
import datetime
YearNow = datetime.date.today().year
age = []
for i in df_mk_camp.index :
    age.append(YearNow - df_mk_camp.at[i,"Year_Birth"])
df_CusInfo.insert(1,"Age", age ) #เพิ่มแถวอายุเข้าไป
```

```python
df_CusInfo
```

```python
df_consumeInfo = df_CusInfo.drop(columns=["Income","Marital_Status" , "Kidhome", "Teenhome"])
df_consumeInfo #ดูเฉพาะวัตถุดิบที่ซื้อ
```

```python
import numpy as np
meat_array = df_consumeInfo['MntMeatProducts'].to_numpy()
fish_array = df_consumeInfo['MntFishProducts'].to_numpy()
fruit_array = df_consumeInfo['MntFruits'].to_numpy()
sweet_array = df_consumeInfo['MntSweetProducts'].to_numpy()
wine_array = df_consumeInfo['MntWines'].to_numpy()
print(fruit_array)
print(type(fruit_array))
```

รายจ่ายรวมจะเป็นตัวแสดงว่า คนคนนั้น ให้ทุนกับวัตถุดิบอาหารเท่าไหร่บ้าง

```python
#หารายจ่ายรวม
total_consu_cost = meat_array + fish_array + fruit_array + sweet_array + wine_array
print(total_consu_cost)
print(type(total_consu_cost))
```

```python
pc_meat = 100*meat_array/total_consu_cost
pc_fish = 100*fish_array/total_consu_cost
pc_fruit = 100*fruit_array/total_consu_cost
pc_sweet = 100*sweet_array/total_consu_cost
pc_wine = 100*wine_array/total_consu_cost
meat_fish = pc_meat + pc_fish #เนื้อและปลารวมกัน
print(type(pc_meat))
print(type(pc_fish))
print(type(pc_fruit))
print(type(pc_sweet))
print(type(pc_wine))
print(type(meat_fish))
```

```python
#เพิ่มลงใน df
df_consumeInfo["Total_Cost"] = total_consu_cost.tolist()
df_consumeInfo["Meat&Fish(%)"] = meat_fish.tolist()
df_consumeInfo["meat(%)"] = pc_meat.tolist()
df_consumeInfo["fish(%)"] = pc_fish.tolist()
df_consumeInfo["fruit(%)"] = pc_fruit.tolist()
df_consumeInfo["sweet(%)"] = pc_sweet.tolist()
df_consumeInfo["wine(%)"] = pc_wine.tolist()
#คิด%ที่จ่ายเมื่อเทียบกับค่าใช้จ่ายด้านโบริดภคอื่นๆ
```

### **ข้อสรุปของการหาลูกค้าที่ควรขายอาหารสุขภาพให้**

จากข้อมูลชุดนี้ พิจารณาว่าลูกค้าคนไหนควรขายอาหารสุขภาพให้
ได้จากหัวข้อต่อไปนี้
1. สินค้าที่ซื้อมีอัตราส่วน ปลากับเนื้อ(โปรตีน) ที่สูง และ ของหวานและไวน์ที่ต่ำ ส่วนผลไม้นั้นควรอยู่ในเกณฑ์ที่ไม่มากจนเกินไปและไม่น้อยจนเกินไป
2. จ่ายเงินมากให้กับการซื้อสินค้ากินดื่ม


---

> อนึ่ง ไวน์นั้น บางครั้งก็มีโอกาสที่จะซื้อไปเพื่อทำการปรุงอาหารด้วย ดังนั้นจึงจัดให้ไวน์อันดับสูงกว่าของหวาน โดยที่หากซื้อเนื้อและปลาในอัตราที่ใกล้กัน ก็สมเหตุสมผล แต่หากไม่ก็มีอาจจะมีแนวโน้มว่าจะซื้อไปเพื่อดื่มเป็นส่วนใหญ่



> แต่เนื่องจากด้วยความที่ไวน์นั้นมีความหลากหลายมาก ทำให้การพิจารณาด้านไวน์นั้นค่อนข้างซับซ้อน เพราะถ้าหากไวน์ที่ซื้อเป็นไวน์เกรดดีหรือไวน์สะสมก็มักจะมีราคาที่สูง จำนวนเงินที่ต้องใช้ ต่อขวด จึงสูง จึงดึงค่า % ให้สูงขึ้นตามไปด้วย แม้จะซื้อไวน์เพียงไม่กี่ขวด
